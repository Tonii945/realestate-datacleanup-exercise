#### Ejercicio 11. Analiza la relación entre la superficie y el precio de las casas. (★★☆)

Pista: Puedes hacer un `scatter plot` y luego escribir una conclusión al respecto.

```python


filtered_data = ds[
    (ds['price'] > 0) & 
    (ds['surface'] > 0) & 
    (ds['surface'] <= 800)
]

plt.figure(figsize=(10, 6))
plt.scatter(
    filtered_data['surface'],
    filtered_data['price'],
    alpha=0.6,
    edgecolor='k',
    s=20, 
    color='blue'
)


plt.title('Relación entre Superficie y Precio de las Casas', fontsize=14)
plt.xlabel('Superficie (m²)', fontsize=12)
plt.ylabel('Precio (USD)', fontsize=12)

import matplotlib.ticker as ticker
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f} USD'))

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()


```

Existe una correlación positiva: a mayor superficie, generalmente, mayor es el precio.

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
sub_ds = sub_ds.copy()  
sub_ds['normalized_price'] = sub_ds.groupby('level5')['price'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)

plt.figure(figsize=(12, 6))

for ciudad in sub_ds['level5'].unique():
    subset = sub_ds[sub_ds['level5'] == ciudad]
    plt.hist(
        subset['normalized_price'], 
        bins=10, 
        alpha=0.5, 
        label=ciudad
    )

plt.title("Histogramas de Precios Normalizados por Población", fontsize=14)
plt.xlabel("Precio Normalizado (0-1)", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.legend(title="Poblaciones")
plt.grid(alpha=0.3)
plt.show()
```

El gráfico puede mostrar si alguna población tiene precios más homogéneos (distribución concentrada) o más diversos (distribución más plana).
Si los histogramas tienen picos similares para diferentes poblaciones, podría indicar patrones comunes en el mercado del cinturón sur de Madrid.

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
sub_ds['pps'] = sub_ds['price'] / sub_ds['surface']

getafe_datos = sub_ds[sub_ds['level5'] == 'Getafe']
alcorcon_datos = sub_ds[sub_ds['level5'] == 'Alcorcón']

pps_getafe = getafe_datos['pps'].mean().round(2)
pps_alcorcon = alcorcon_datos['pps'].mean().round(2)

# Imprimir resultados
print("Estadísticas de Precio por Metro Cuadrado:")
print("\nGetafe:")
print(pps_getafe)
print("\nAlcorcón:")
print(pps_alcorcon)


print("Los resultados indican que Alcorcón tiene un mercado inmobiliario más caro que Getafe.\nPara una inversión, Getafe podría ser más económico, mientras que Alcorcón podría representar mayor rentabilidad a largo plazo debido a su mercado más competitivo.")
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)
